# Gradient Boosting Regression


```python
from sklearn.metrics import r2_score, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.datasets import make_classification
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
```

```python
# Generate Data
x, y = make_regression(n_samples=1000, n_features=10, noise=1, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

```python
GBR = GradientBoostingRegressor(n_estimators=299, learning_rate=0.01, random_state=42, max_depth=4, subsample=0.8)
# subsample is used to define with what percentage from the total sample data should be fed to the weak learners (each Decision Tree)
GBR.fit(x_train, y_train)
pred_y= GBR.predict(x_test)
print("The r2 score of GBR is :", r2_score(y_test, pred_y))
```

# Gradient Boosting Classification


```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=15, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)

GBC = GradientBoostingClassifier(n_estimators=299, learning_rate=0.01, random_state=42)
GBC.fit(x_train, y_train)
predc_y= GBC.predict(x_test)
print("The accuracy score of GBR is :", accuracy_score(y_test, predc_y))
```

# Adaboost classification


```python
x, y = make_classification(n_samples=1000, n_features=20, n_informative=15, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
base_model = DecisionTreeClassifier(max_depth= 2)
ADC = AdaBoostClassifier(
    estimator=base_model,
    n_estimators=200, 
    random_state=42)
ADC.fit(x_train,y_train)
preda_y = ADC.predict(x_test)
print("The accuracy score of ADC is :", accuracy_score(y_test, preda_y))
```

# VOTING

```python
from sklearn.metrics import r2_score, classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import VotingRegressor
from sklearn.datasets import make_classification
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.svm import SVR
```

```python
x, y = make_classification (
    n_samples=2000,
    n_features=35,
    n_redundant=15,
    n_informative=15,
    random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
svc = SVC()
lr = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=3)
vot = VotingClassifier(
    [("dt", dt),
     ("lr", lr),
     ("svc", svc)])
vot.fit(x_train, y_train)
pred_vc = vot.predict(x_test)
print("The predicted result of the Voting classifer are", classification_report(y_test, pred_vc))
print("The accuracy result of the Voting classifer are", accuracy_score(y_test, pred_vc))
```

# Voting Regressor

```python
x, y = make_regression (
    n_samples=2000,
    n_features=35,
    n_informative=15,
    random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
dtr = DecisionTreeRegressor()
lr = LinearRegression()
svr = SVR()
vot = VotingRegressor(
    [("dtr", dtr),
     ("lr", lr),
     ("svr", svr)])
vot.fit(x_train, y_train)
pred_vc = vot.predict(x_test)
print("The r2 result of the Voting classifer are", r2_score(y_test, pred_vc))
```

# STACKING CLASSIFIER

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import StackingRegressor
x, y = make_classification (
    n_samples=2000,
    n_features=35,
    n_redundant=15,
    n_informative=15,
    random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
final_esti = LogisticRegression()
svc = SVC()
lr = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=3)
sta = StackingClassifier(
    [("dt", dt),
     ("lr", lr),
     ("svc", svc)],
final_estimator=final_esti,
    cv=5,)
    
    
sta.fit(x_train, y_train)
pred_vc = sta.predict(x_test)
print("The predicted result of the Voting classifer are", classification_report(y_test, pred_vc))
print("The accuracy result of the Voting classifer are", accuracy_score(y_test, pred_vc))
```

# ASSIGNMENT

```python
from sklearn.metrics import r2_score, classification_report, accuracy_score, recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = pd.read_csv("novagen_dataset.csv")
print(data.columns)
data
```

```python
# label the boolen values 
lb = LabelEncoder()
data["Diet_Type__Vegan"] = lb.fit_transform(data["Diet_Type__Vegan"])
data["Diet_Type__Vegetarian"] = lb.fit_transform(data["Diet_Type__Vegetarian"])
data["Blood_Group_AB"] = lb.fit_transform(data["Blood_Group_AB"])
data["Blood_Group_B"] = lb.fit_transform(data["Blood_Group_B"])
data["Blood_Group_O"] = lb.fit_transform(data["Blood_Group_O"])
# split the data

x = data.drop("Target", axis =1)
y = data["Target"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42,stratify=y)
ls = StandardScaler()
x_test_scaled = ls.fit_transform(x_test)
x_train_sclaed = ls.fit_transform(x_train)
```

# KNN MODEL

```python
knn = KNeighborsClassifier( n_neighbors=5,
    metric="euclidean")
knn.fit(x_train_sclaed, y_train)
pred_k = knn.predict(x_test_scaled)
print("The accuracy of the model is ", accuracy_score(y_test, pred_k))
print("The recall_score of the model is ", recall_score(y_test, pred_k))
```

# Logistic Regression with Regularization 

```python
LR = LogisticRegression( penalty="l2",
    solver="liblinear",
    max_iter=1000)
LR.fit(x_train_sclaed, y_train)
pred_LR = LR.predict(x_test_scaled)
print("The accuracy of the model is ", accuracy_score(y_test, pred_LR))
print("The recall_score of the model is ", recall_score(y_test, pred_LR))
```

# Random Forest

```python
RF = RandomForestClassifier( n_estimators=200,
    max_depth=None,
    random_state=42)
RF.fit(x_train_sclaed, y_train)
pred_RF = RF.predict(x_test_scaled)
print("The accuracy of the model is ", accuracy_score(y_test, pred_RF))
print("The recall_score of the model is ", recall_score(y_test, pred_RF))
```

# Gradient BOOSTING Classifier

```python
GDC = GradientBoostingClassifier( n_estimators=200,
    max_depth=None,
    random_state=42)
GDC.fit(x_train_sclaed, y_train)
pred_GDC= GDC.predict(x_test_scaled)
print("The accuracy of the model is ", accuracy_score(y_test, pred_GDC))
print("The recall_score of the model is ", recall_score(y_test, pred_GDC))
```

# VOTING CLASSIFIER


```python
VC = VotingClassifier(estimators=[("lr", LogisticRegression(max_iter=1000, solver="liblinear")),
                        ("knn", KNeighborsClassifier(n_neighbors=5)),
                         ("rf", RandomForestClassifier(n_estimators=200, random_state=42))],
                          voting = "soft"
                       )
VC.fit(x_train_sclaed, y_train)
pred_VC= VC.predict(x_test_scaled)
print("The accuracy of the model is ", accuracy_score(y_test, pred_VC))
print("The recall_score of the model is ", recall_score(y_test, pred_VC))
```
