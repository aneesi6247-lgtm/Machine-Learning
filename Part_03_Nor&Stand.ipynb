```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# importing and spliting the data
data = pd.read_csv("heart.csv")
x = data.drop("target", axis = 1)
y = data["target"]
x_train, x_test, y_train, y_test = train_test_split(
    x, y, random_state=42, test_size=0.2)

#standarization
stand = StandardScaler()
x_train_scaled = stand.fit_transform(x_train)
x_test_scaled = stand.transform(x_test)

model= LogisticRegression(max_iter=1000)
model.fit(x_train_scaled,y_train)
predy = model.predict(x_test_scaled)
print("The accuracy of the model is", accuracy_score(y_test, predy))
print("The Precision of the model is", precision_score(y_test, predy))

```

```python
cm = confusion_matrix(y_test, predy)
print(cm)
report = classification_report(y_test, predy)
print(report)
```

```python
## Naive Bayes Regression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, accuracy_score, f1_score, recall_score
naive = GaussianNB()
naive.fit(x_train, y_train)
naiv_pre = naive.predict(x_test)
print("The Precision value is :", precision_score(y_test, naiv_pre))
print("The Accuracy value is :", accuracy_score(y_test, naiv_pre))
print("The  F1_score value is :", f1_score(y_test, naiv_pre))
print("The Recall_score value is :", recall_score(y_test, naiv_pre))
```

```python
# KNN - K Nearest Neighbours
# For knn we must used scaled values
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)
knn_pred = knn.predict(x_test_scaled)
print("The Precision value is :", precision_score(y_test, knn_pred))
print("The Accuracy value is :", accuracy_score(y_test, knn_pred))
print("The  F1_score value is :", f1_score(y_test, knn_pred))
print("The Recall_score value is :", recall_score(y_test, knn_pred))
```

```python
# for K = 7
# this the best value for recall score
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train_scaled, y_train)
knn_pred = knn.predict(x_test_scaled)
print("The Precision value is :", precision_score(y_test, knn_pred))
print("The Accuracy value is :", accuracy_score(y_test, knn_pred))
print("The  F1_score value is :", f1_score(y_test, knn_pred))
print("The Recall_score value is :", recall_score(y_test, knn_pred))
```

```python
# for k = 9
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train_scaled, y_train)
knn_pred = knn.predict(x_test_scaled)
print("The Precision value is :", precision_score(y_test, knn_pred))
print("The Accuracy value is :", accuracy_score(y_test, knn_pred))
print("The  F1_score value is :", f1_score(y_test, knn_pred))
print("The Recall_score value is :", recall_score(y_test, knn_pred))
```

```python
# Cross validation for for hyperperameters tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
prag_grid = {"n_neighbors" : [3, 5, 7, 9, 11]}
classifier_cv = GridSearchCV(
    classifier,
    prag_grid,
    cv = 5)
classifier_cv.fit(x_train_scaled, y_train)
out= classifier_cv.predict(x_test_scaled)
print("The Precision value is :", precision_score(y_test, out))
print("The Accuracy value is :", accuracy_score(y_test, out))
print("The  F1_score value is :", f1_score(y_test, out))
print("The Recall_score value is :", recall_score(y_test, out))
# results
result = pd.DataFrame(classifier_cv.cv_results_)
print(result[["param_n_neighbors", "mean_test_score"]])
print(classifier_cv.best_params_)
```

```python
from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
prag_grid = {"n_neighbors" : [3, 5, 7, 9, 11]}
classifier_cv = GridSearchCV(
    classifier,
    prag_grid,
    cv = 5,
    scoring = "recall")
# scoring is defined to get the results based on a particular evaluation matrics for example, accuracy, recall, or precision
classifier_cv.fit(x_train, y_train)
out= classifier_cv.predict(x_test)
print("The Precision value is :", precision_score(y_test, out))
print("The Accuracy value is :", accuracy_score(y_test, out))
print("The  F1_score value is :", f1_score(y_test, out))
print("The Recall_score value is :", recall_score(y_test, out))
# results
result = pd.DataFrame(classifier_cv.cv_results_)
print(result[["param_n_neighbors", "mean_test_score"]])
print(classifier_cv.best_params_)
```

```python
# Pipelines
# scaling is learned only from the training data, so no data leakage,
# Everthing in one object
# Easy tuning of Hyperparameters

from sklearn.pipeline import Pipeline
x_train, x_test, y_train, y_test = train_test_split(
    x, y, random_state=42, test_size=0.2)
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])
prag_grid = {"knn__n_neighbors" : [3, 5, 7, 9, 11]}
classifier_cv = GridSearchCV(
    pipeline,
    prag_grid,
    cv = 5,
    scoring = "recall")
classifier_cv.fit(x_train, y_train)
out= classifier_cv.predict(x_test)
print("The Precision value is :", precision_score(y_test, out))
print("The Accuracy value is :", accuracy_score(y_test, out))
print("The  F1_score value is :", f1_score(y_test, out))
print("The Recall_score value is :", recall_score(y_test, out))
```

```python
## Assignment related to the Iris Dataset
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
```

```python
iris_data = pd.read_csv("Iris.csv")
iris_data.head()
# we have to conver the species into 0,1, 2
ld = LabelEncoder()
iris_data["Species"] = ld.fit_transform(iris_data["Species"])

# defining the x and y
x = iris_data.drop("Species", axis = 1)
y = iris_data["Species"]


# spliting the data in x and y
xa_train, xa_test, ya_train, ya_test = train_test_split(
    x, y, test_size=0.5, random_state=42)
```

```python
# Logistic Regression
lgreg = LogisticRegression(max_iter=1000)
lgreg.fit(xa_train, ya_train)
pred_LG= lgreg.predict(xa_test)
print("The Precision value is :", precision_score(ya_test, pred_LG, average='macro'))
print("The Accuracy value is :", accuracy_score(ya_test, pred_LG))
print("The  F1_score value is :", f1_score(ya_test, pred_LG,average='macro'))
print("The Recall_score value is :", recall_score(ya_test, pred_LG,average='macro'))
#* macro → treats all classes equally (best for understanding model quality)
#weighted → considers class imbalance
#micro → overall global performance
```

```python
# KNN Neighbors
knn_n = KNeighborsClassifier()
knn_n.fit(xa_train, ya_train)
pred_knn = knn_n.predict(xa_test)
print("The Precision value is :", precision_score(ya_test, pred_knn, average='macro'))
print("The Accuracy value is :", accuracy_score(ya_test, pred_knn))
print("The  F1_score value is :", f1_score(ya_test, pred_knn,average='macro'))
print("The Recall_score value is :", recall_score(ya_test, pred_knn,average='macro'))
```

```python
# for Naive Bayes
nb = GaussianNB()
nb.fit(xa_train, ya_train)
pred_NA = nb.predict(xa_test)
print("The Precision value is :", precision_score(ya_test, pred_NA, average='macro'))
print("The Accuracy value is :", accuracy_score(ya_test, pred_NA))
print("The  F1_score value is :", f1_score(ya_test, pred_NA,average='macro'))
print("The Recall_score value is :", recall_score(ya_test, pred_NA,average='macro'))
```
